```python
df=df.drop(labels=['id','date','waterfront','view','yr_built','yr_renovated','zipcode','lat','long','sqft_living15','sqft_lot15','condition','grade'],axis=1)
```

```python
for i in df.columns:
    print(f"Column : {i} and values are : {df[i].head().tolist()}")
```

```python
df=df.dropna(axis=0)
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
t1 = df.drop(['price'], axis=1)
feat = df['price']

X_train, X_test, y_train, y_test = train_test_split(t1, feat, test_size=0.30, random_state=52)

model=LinearRegression()

model.fit(X_train,y_train)

s=model.predict(X_test)

mae = mean_absolute_error(y_test,s)
mse = mean_squared_error(y_test,s)
rmse = mean_squared_error(y_test,s, squared=False)
r2 = r2_score(y_test,s)

print(f"""
Mean Absolute error : {mae}
Mean Squared Error : {mse}
Root mean squared error : {rmse}
Rsqaure Score : {r2}
""")
```

```python
import numpy as np
import pandas as pd
np.random.seed(42)
sam = 20000

bed = np.random.randint(1, 8, sam)
bath = np.round(np.random.uniform(1, 5, sam), 0)
liv = np.random.randint(500, 10000, sam)
lot = np.random.randint(1000, 15000, sam)
flo = np.random.choice([1, 2, 3,4], sam)
abo = np.random.randint(500, 8000, sam)
base = liv - abo
base[base < 0] = 0 

price = (
    10000 + 50000 * bed +
    30000 * bath +
    150 * liv +
    50 * lot +
    10000 * flo +
    200 * abo +
    100 * base +
    np.random.normal(0, 100000, sam)
)

data = {
    'price': price,
    'bedrooms': bed,
    'bathrooms': bath,
    'sqft_living': liv,
    'sqft_lot': lot,
    'floors': flo,
    'sqft_above': abo,
    'sqft_basement': base
}

trainer = pd.DataFrame(data)

print(trainer.head())

trainer.to_csv('synthetic_house_data.csv', index=False)
```
